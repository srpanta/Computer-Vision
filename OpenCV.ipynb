```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
import cv2
```

```python
img = cv2.imread("Denali.JPG")
if img is None:  
    print("Could not open or find the image")
type(img)
```

```python
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
```

```python
type(img)
```

```python
plt.imshow(img)
```

## There is a small animal at the middle. Let's put a circle around this animal.

```python
cv2.circle(img, center=(3000,2100), radius=200,thickness=40,color=(0,0,255))
plt.imshow(img)
```

## Cropping an image.
### To make the animal more visible, middle portion of the image can be cropped and magnified.

```python
cropped_image = img[1500:2500,2600:3500]
plt.imshow(cropped_image)
```

## Let's insert a text.
### Text can be inserted (if necessary) inside the image.

```python
font = cv2.FONT_HERSHEY_SIMPLEX
cv2.putText(cropped_image,text="MOOSE inside the circle",org=(100,280), fontFace=font,fontScale=2,
            color=(255,0,0),thickness=4,lineType=cv2.LINE_AA)
plt.imshow(cropped_image)
```

## Let's put this text inside a rectangular box.

```python
cv2.rectangle(cropped_image, pt1=(80,300),pt2 = (850,200),color=(0,0,255), thickness=10)
plt.imshow(cropped_image)
```

```python
cropped_image.shape
```

## Resize the image

```python
plt.figure(figsize=(10,10))
wratio=1.2 #increase width 
hratio=0.9 #reduce height
new_img = cv2.resize(cropped_image,(0,0), cropped_image,wratio,hratio)
plt.imshow(new_img)
```

```python
new_img.shape,cropped_image.shape
```

```python
## If we like to flip the image.
flipped_img=cv2.flip(new_img,0)# argument 0 for verticla flip and 1 for horizontal flip
plt.imshow(flipped_img)
```

## Writing an image file

```python
cv2.imwrite("image_moose.jpg",new_img)
```
